```python
import pandas as pd
import re

# 1. ПОЛНЫЙ СПИСОК ПРИЗНАКОВ (Все 33+)
items_to_extract = {
    'Бочонки': 'бочон',
    'Фишки': 'фиш', # Укоротили корень для надежности
    'Жетоны': 'жетон',
    'Маркеры': 'маркер',
    'Кубики': 'кубик|зары',
    'Инструкция': 'инструкц',
    'Игровое поле': 'игров\w*\s+поле',
    'Фломастер': 'фломастер',
    'Карточки': 'карточк',
    'Мешочки': 'мешоч|мешок|мешк',
    'Коробка': 'коробк',
    'Правила игры': 'правил',
    'Игральные кости': 'кост',
    'Фигурки': 'фигур',
    'Колода': 'колод',
    'Карты': 'карт', 
    'Значки': 'знач',
    'Детали': 'детал',
    'Элементы': 'элемент',
    'Стрелки': 'стрелк',
    'Шарики': 'шарик',
    'Брошюры': 'брошюр',
    'Плакат': 'плакат',
    'Стаканы': 'стакан',
    'Песочные часы': 'песоч\w*\s+час|час\w*\s+песоч|песоч',
    'Деньги': 'деньг|купюр',
    'Памятка': 'памятк',
    'Бруски': 'брусок|брусоч',
    'Камни': 'камни|камен',
    'Миниатюры': 'миниатюр',
    'Ящик': 'ящик',
    'Органайзер': 'органайзер',
    'Задания': 'задани'
}

# 2. ФУНКЦИЯ ОЧИСТКИ ТЕКСТА
def clean_text_for_search(text):
    # Заменяем латинские буквы-двойники на кириллицу (на всякий случай)
    lat_to_cyr = {'a': 'а', 'e': 'е', 'o': 'о', 'p': 'р', 'c': 'с', 'x': 'х', 'y': 'у', 'k': 'к'}
    text = str(text).lower()
    for lat, cyr in lat_to_cyr.items():
        text = text.replace(lat, cyr)
    # Унифицируем разделители
    text = text.replace(',', ';').replace('\n', ';').replace('.', ';')
    return text

# 3. УНИВЕРСАЛЬНАЯ ФУНКЦИЯ ПОДСЧЕТА
def extract_and_sum_all(text, item_name, search_pattern):
    if pd.isna(text) or str(text).strip() in ["0", ""]:
        return 0
    
    text = clean_text_for_search(text)
    parts = text.split(';')
    
    total_sum = 0
    found_any = False
    
    for part in parts:
        part = part.strip()
        if not part: continue
        
        # ЛОГИКА РАЗДЕЛЕНИЯ КАРТ И КАРТОЧЕК
        if item_name == 'Карты':
            if 'карточк' in part: continue
            is_match = 'карт' in part
        elif item_name == 'Карточки':
            is_match = 'карточк' in part
        else:
            # Для всех остальных (включая Фишки)
            is_match = re.search(search_pattern, part) is not None

        if is_match:
            found_any = True
            
            # А) Удаляем всё в скобках: (5,5х8,2 см)
            part = re.sub(r'\(.*?\)', '', part)
            # Б) Удаляем размеры типа "5х5", "5.5", "5,5"
            part = re.sub(r'\d+\s*[xх*.,]\s*\d+', '', part)
            
            # В) Ищем все числа в блоке
            numbers = re.findall(r'\d+', part)
            
            if numbers:
                total_sum += sum(int(n) for n in numbers)
            else:
                total_sum += 1
                
    return total_sum if found_any else 0

# 4. ПРИМЕНЕНИЕ КО ВСЕМУ ДАТАФРЕЙМУ
for col, pattern in items_to_extract.items():
    df_clean_2[col] = df_clean_2['Комплектация'].apply(lambda x: extract_and_sum_all(x, col, pattern))


```

```python
MATERIAL_MAP = {
    "дерево": [
        "дерево", "фанера", "дуб", "бук", "ясень", "береза", "сосна",
        "массив", "шпон", "хдф", "мдф", "двп", "липа", "орех", "граб"
    ],
    "пластик": [
        "пластик", "abs", "abc", "pp", "pvc", "пвх", "eva",
        "полимер", "полипропилен", "полистирол", "пэт", "pet",
        "акрил", "оргстекло", "пластмасса"
    ],
    "картон": [
        "картон", "мелованный", "ламинированный", "дизайнерский",
        "крафт"
    ],
    "бумага": [
        "бумага", "целлюлоза", "мелованная бумага", "крафт бумага"
    ],
    "ткань": [
        "ткань", "текстиль", "фетр", "бархат", "велькро",
        "липучк", "экокожа", "кожа"
    ],
    "металл": [
        "металл", "алюминий", "жесть", "железо", "бронза",
        "цинк", "сталь"
    ],
    "магнит": [
        "магнит", "магнитный", "магнитная лента", "магнитный винил"
    ],
    "стекло": [
        "стекло"
    ],
    "резина": [
        "резина", "силикон", "эластич"
    ]
}
```

```python
import re

def canonicalize_material(text):
    if pd.isna(text):
        return ""

    text = str(text).lower()

    found = set()

    for canonical, keywords in MATERIAL_MAP.items():
        for kw in keywords:
            if kw in text:
                found.add(canonical)

    return "; ".join(sorted(found))
```

```python
import pandas as pd
import re

def clean_player_data(value):
    if pd.isna(value) or value == '':
        return None, None
    
    val_str = str(value).lower()
    
    # Обработка бесконечности
    if 'не ограничено' in val_str:
        return 1, 99
    
    # Извлечение всех чисел
    nums = [int(n) for n in re.findall(r'\d+', val_str)]
    
    if not nums:
        return None, None
    
    # Возвращаем (минимум, максимум)
    return min(nums), max(nums)

# 1. Генерируем числовые столбцы из "объекта"
temp_data = wb_final['Макс. число игроков'].apply(clean_player_data)
wb_final['min_players'] = temp_data.apply(lambda x: x[0])
wb_final['max_players'] = temp_data.apply(lambda x: x[1])

# 2. Заполняем пропуски медианой и приводим к целому числу
wb_final['min_players'] = wb_final['min_players'].fillna(wb_final['min_players'].median()).astype(int)
wb_final['max_players'] = wb_final['max_players'].fillna(wb_final['max_players'].median()).astype(int)

# Удаляем старую текстовую колонку, чтобы она не дублировала данные
# wb_final.drop(columns=['Макс. число игроков'], inplace=True)

print(wb_final[['min_players', 'max_players']].info())
```
